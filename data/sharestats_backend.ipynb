```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame
df_sample = df[['pmcid', 'doi', 'journal_title', 'title', 'journal_year', 'open_data', 'data_share', 'data_score', 'project_id', 'contact_pi_project_leader', 'organization_name']]

# Identify rows with the same 'pmcid' but different 'contact_pi_project_leader'
differing_leaders_mask = df_sample.duplicated(subset=['pmcid'], keep=False) & (df_sample.duplicated(subset=['pmcid', 'contact_pi_project_leader'], keep=False))

# Extract a sample of rows with the specified condition
sample_rows_differing_leaders = df_sample[differing_leaders_mask].sample(n=500)

# Extract a sample of truly random rows
sample_rows_random = df_sample[~differing_leaders_mask].sample(n=500)

# Combine the two samples
final_sample = pd.concat([sample_rows_differing_leaders, sample_rows_random])

# Display the final sample
print("Final Sample of Rows:")
final_sample
```
